### Student Academics Performance 
Fonte: https://archive.ics.uci.edu/ml/datasets/Student+Academics+Performance#

```python
student_data = arff.loadarff('Sapfile1.arff')
student = pd.DataFrame(student_data[0])
student.head()
```

```python
%matplotlib inline
student.atd.value_counts().plot(kind="bar")
```

```python
student.shape
```

### Financial Distress
Fonte: https://www.kaggle.com/shebrahimi/financial-distress

```python
distress = pd.read_csv('Financial Distress.csv')
distress['distress'] = distress['Financial Distress'].apply(lambda x: 1 if x<-0.5 else 0)
distress.head()
```

```python
distress['distress'].value_counts().plot(kind='bar')
```
